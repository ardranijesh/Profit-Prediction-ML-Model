As this task is based on the problem of regression so I will be using the Linear regression algorithm to train the profit prediction model. So let’s prepare the data so that we can fit it into the model:

```python
features_df= df[["R&D Spend", "Administration", "Marketing Spend"]]
target_df= df["Profit"]
```

```python
from sklearn.model_selection import train_test_split

features_train, features_val, target_train, target_val = train_test_split(features_df, target_df, test_size=0.2, random_state=100)
```

# *Linear Regression*

```python
LR_model = LinearRegression()
LR_model.fit(features_train, target_train)
predictions = LR_model.predict(features_train)
predictions_val = LR_model.predict(features_val)
compare = pd.DataFrame({'Actuals' : target_train, 'predictions': predictions, 'diffrence': np.absolute(target_train-predictions)})
compare_val = pd.DataFrame({'Actuals': target_val, 'predictions': predictions_val, 'diffrence': np.absolute(target_val-predictions_val) })
print(compare_val)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(target_val, predictions_val)
print("Mean Absolute Error is :" ,mae)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(target_val, predictions_val)
print("Mean Squarred Error is :" ,mse*100)
rmse = np.sqrt(mean_squared_error(target_val, predictions_val))
print("Root Mean Squarred Error is : ",rmse*100)
from sklearn.metrics import r2_score
r2Score = r2_score(target_val, predictions_val)
print("R2 score of model is :" ,r2Score*100)
```

# *Decision Tree Regressor*

```python
from sklearn.tree import DecisionTreeRegressor
clr = DecisionTreeRegressor()
clr.fit(features_train, target_train)
predictions = clr.predict(features_train)
predictions_val = clr.predict(features_val)
compare = pd.DataFrame({'Actuals' : target_train, 'predictions': predictions, 'diffrence': np.absolute(target_train-predictions)})
compare_val = pd.DataFrame({'Actuals': target_val, 'predictions': predictions_val, 'diffrence': np.absolute(target_val-predictions_val) })
print(compare_val)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(target_val, predictions_val)
print("Mean Absolute Error is :" ,mae)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(target_val, predictions_val)
print("Mean Squarred Error is :" ,mse*100)
rmse = np.sqrt(mean_squared_error(target_val, predictions_val))
print("Root Mean Squarred Error is : ",rmse*100)
from sklearn.metrics import r2_score
r2Score = r2_score(target_val, predictions_val)
print("R2 score of model is :" ,r2Score*100)
```
